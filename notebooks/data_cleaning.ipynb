## Exploracion de la data

```python
# imports
import os
import pandas as pd
```

```python
# Ruta a la carpeta de data
DATA_PATH = "../data"

# Listar archivos disponibles en la carpeta de datos
csv_files = [f for f in os.listdir(DATA_PATH) if f.endswith(".csv")]
print("Archivos disponibles:", csv_files)

# FunciÃ³n para explorar un CSV
def explorar_csv(file_name):
    print(f"\nðŸ“Œ Explorando: {file_name}")
    df = pd.read_csv(os.path.join(DATA_PATH, file_name))
    print("ðŸ”¹ Dimensiones:", df.shape)
    display(df.head())  # Para Jupyter Notebook
    print("ðŸ”¹ Tipos de datos:\n", df.dtypes)
    print("ðŸ”¹ Valores nulos:\n", df.isnull().sum())

# Explorar cada archivo
for file in csv_files:
    explorar_csv(file)
```

## Data Cleaning

```python
# Cargar los archivos CSV a limpiar
constructor_results = pd.read_csv('../data/constructor_results.csv')
drivers = pd.read_csv('../data/drivers.csv')
qualifying = pd.read_csv('../data/qualifying.csv')
races = pd.read_csv('../data/races.csv')
```

```python
# Reemplazar \N por NaN en las columnas relevantes
constructor_results.replace(r'\\N', pd.NA, regex=True, inplace=True)
drivers.replace(r'\\N', pd.NA, regex=True, inplace=True)
qualifying.replace(r'\\N', pd.NA, regex=True, inplace=True)
```

```python
# Convertir la columna 'number' de drivers a int64, manejando NaN
drivers['number'] = pd.to_numeric(drivers['number'], errors='coerce', downcast='integer')
```

```python
# Verificar duplicados en todos los archivos y eliminarlos si es necesario
constructor_results.drop_duplicates(inplace=True)
drivers.drop_duplicates(inplace=True)
qualifying.drop_duplicates(inplace=True)
races.drop_duplicates(inplace=True)
```

##### Revisar las Fechas en `races.csv`
Asegurarse de que las columnas `date` y `time` en `races.csv` estÃ©n correctamente tipificadas. Para eso se puede convertirlas al tipo datetime.

```python
# Convertir las columnas de fecha y hora a tipo datetime con formato especÃ­fico
races['date'] = pd.to_datetime(races['date'], format='%Y-%m-%d', errors='coerce')
races['time'] = pd.to_datetime(races['time'], format='%H:%M:%S', errors='coerce')
```

```python
# Guardar los archivos limpios
constructor_results.to_csv('../data/constructor_results_cleaned.csv', index=False)
drivers.to_csv('../data/drivers_cleaned.csv', index=False)
qualifying.to_csv('../data/qualifying_cleaned.csv', index=False)
races.to_csv('../data/races_cleaned.csv', index=False)
```
